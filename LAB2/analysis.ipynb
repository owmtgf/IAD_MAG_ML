```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./dataset/train.csv", index_col=0)
df["Sentence_id"] = df["Sentence_id"].ffill()
df.head(30)
```

```python
print(sorted(df["Tag"].unique().tolist()))
print(df["Tag"].value_counts())
```

```python
def print_unique_values(df):
    """Print unique values for each column"""

    for col in df.columns:
        unique_vals = df[col].unique()
        print(f"\nColumn: {col}")
        print(f"Unique values count: {len(unique_vals)}")

        # print first 20 values to avoid huge output
        print(unique_vals[:20])

print_unique_values(df)
```
